# Construction d'un modèle 

## Modèle de Poisson

### Création du modèle

Essayons de prédire la dernière journée de championnat. 

```python
L1_1920_train = L1_1920[:-10]
```

```python
L1_1920_test = L1_1920[-10:]
```

```python
L1_1920_test
```

```python
# importing the tools required for the Poisson regression model
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
modele_buts_data = pd.concat([L1_1920_train[['HomeTeam','AwayTeam','HomeGoals']].assign(home=1).rename(
            columns={'HomeTeam':'team', 'AwayTeam':'opponent','HomeGoals':'goals'}),
           L1_1920_train[['AwayTeam','HomeTeam','AwayGoals']].assign(home=0).rename(
            columns={'AwayTeam':'team', 'HomeTeam':'opponent','AwayGoals':'goals'})])
```

```python
modele_buts_data.head()
```

```python
poisson_model = smf.glm(formula="goals ~ home + team + opponent", data=modele_buts_data, 
                        family=sm.families.Poisson()).fit()
print(poisson_model.summary())
```

L'intérprétation des coefficients d'un modèle de Poisson se fait de la même manière que pour un modèle de régression logistique, i.e on prend l'exponentielle du coefficient. Un coefficient positif ($e^{x}>1  \forall x > 0$) implique une probabilité de buts plus élevé tandis qu'un coefficient proche de 0 ($e^{0}=1$) indique un effet plus neutre. Prenons un exemple pour expliciter les coefficients. Prenons par exemple la dernière ligne de notre tableau de coefficients : 'home'. 'home' a un coefficient égal à 0.3477. Le signe positif du coefficient nous confirme l'intuition que l'on pouvait avoir que les équipes à domicile marquent en moyenne plus de buts que les équipes à l'extérieur. En particulier, c'est $e^{0.3477}=1.42$ fois plus probable que les équipes à domicile marquent plus de buts que les équipes à l'extérieur. 
D'autre part, nous voyons que le coefficient team associé au Paris SG est positif tandis que celui de Toulouse est négatif. Cela nous indique que le Paris SG (resp. Toulouse) marque en moyenne plus de buts (moins de buts) que la moyenne des équipes de Ligue 1. 
Enfin, les coefficients associés à opponent nous renseigne sur la force défensive des équipes. Nous voyons notamment que le coefficient associé au PSG (Toulouse) est négatif (positif) ce qui signifie que le PSG encaisse en moyenne relativement moins de buts (plus de buts) que les autres équipes du championnat. 

### Prédiction : Exemple sur un match

```python
def simulate_match(modele_foot, EquipeDom, EquipeExt, max_goals=10):
    home_goals_avg = modele_foot.predict(pd.DataFrame(data={'team': EquipeDom,
                                                           'opponent': EquipeExt, 'home': 1},
                                                     index=[1])).values[0]
    away_goals_avg = modele_foot.predict(pd.DataFrame(data={'team': EquipeExt,
                                                           'opponent': EquipeDom, 'home': 0},
                                                     index=[1])).values[0]
    team_pred = [[poisson.pmf(i, team_avg) for i in range(
        0, max_goals+1)] for team_avg in [home_goals_avg, away_goals_avg]]
    return(np.outer(np.array(team_pred[0]), np.array(team_pred[1])))
```

```python
print('Notre modèle prédit que Lyon marquera {} buts lors de ce match.'.format(round(poisson_model.predict(pd.DataFrame(data={'team': 'Lyon', 'opponent': 'Reims',
                                                                                                                              'home': 1}, index=[1]))[1], 2)))
print('Notre modèle prédit que Reims marquera {} buts lors de ce match.'.format(round(poisson_model.predict(pd.DataFrame(data={'team': 'Reims', 'opponent': 'Lyon',
                                                                                                                               'home': 0}, index=[1]))[1], 2)))
```

```python
lyon_reims = simulate_match(poisson_model, 'Lyon', 'Reims', max_goals=10)
```

```python
print("Match : Lyon vs Reims :")
print("Victoire de Lyon : {}%".format(round(np.sum(np.tril(lyon_reims, -1))*100, 2)))
print("Victoire de Reims : {}%".format(round(np.sum(np.triu(lyon_reims, 1))*100, 2)))
print("Match Nul : {}%".format(round(np.sum(np.diag(lyon_reims))*100,2)))
```

```python
from matplotlib.colors import ListedColormap


def matrix_gif(matrix, colour_matrix, colour_map, subtitle="", heatmap=False, alpha=0.8):
    fig, ax1 = plt.subplots(1, figsize=(5, 5))
    if heatmap:
        ax1.matshow(matrix, alpha=alpha)
    else:
        ax1.matshow(colour_matrix, cmap=colour_map, alpha=alpha)
    ax1.tick_params(axis=u'both', which=u'both', length=0)
    ax1.grid(which='major', axis='both', linestyle='')
    ax1.set_xlabel("Buts Equipe Extérieur", fontsize=12)
    ax1.set_ylabel('Buts Equipe Domicile', fontsize=12)
    ax1.xaxis.set_label_position('top')
    nrows, ncols = matrix.shape
    for i in range(nrows):
        for j in range(ncols):
            c = matrix[i][j]
            ax1.text(j, i, str(round(c, 4)), va='center', ha='center', size=13)
    plt.figtext(0.5, 0.05, subtitle, horizontalalignment='center',
                fontsize=14, multialignment='left', fontweight='bold')
    return fig


cmap = ListedColormap(['w', '#04f5ff', '#00ff85', '#e90052'])
matrix = simulate_match(poisson_model, 'Lyon', 'Reims', max_goals=5)
matn = len(matrix)
matrix_gif(matrix, matrix, ListedColormap(['w']), heatmap=True,
           alpha=0.6, subtitle="Matrice de probabilité de score du match").savefig("match_matrix_0.png")
```

```python
for t, (mat, colour, subtitle) in enumerate(zip([np.zeros((matn, matn)), np.tril(np.ones((matn, matn)), -1),
                                                 np.triu(
                                                     np.ones((matn, matn))*2, 1), np.diag([3]*matn)],
                                                ['w', '#04f5ff', '#00ff85',
                                                    '#e90052', '#EAF205'],
                                                ['Matrice de probabilité du score du match', "Victoire de l'équipe à domicile",
                                                 "Victoire de l'équipe à l'extérieur", 'Match Nul'])):
    matrix_gif(matrix, mat, ListedColormap(['w'] + [colour]), heatmap=False,
               alpha=0.6, subtitle=subtitle).savefig("match_matrix_{}.png".format(t+1))
```

### Prédiction sur l'ensemble de la journée

```python
buts_home_pred = []
buts_away_pred = []
home_team = []
away_team = []
prob_domicile = []
prob_exterieur = []
prob_nuls = []
for i, j in zip(L1_1920_test['HomeTeam'], L1_1920_test['AwayTeam']):
    match = simulate_match(poisson_model, '{}'.format(i),
                           '{}'.format(j), max_goals=10)
    home_team.append(i)
    away_team.append(j)
    prob_domicile.append(round(np.sum(np.tril(match, -1))*100, 2))
    prob_exterieur.append(round(np.sum(np.triu(match, 1))*100, 2))
    prob_nuls.append(round(np.sum(np.diag(match))*100, 2))
    buts_home_pred.append(poisson_model.predict(pd.DataFrame(data={'team': '{}'.format(i), 'opponent': '{}'.format(j),
                                                                   'home': 1}, index=[1]))[1])
    buts_away_pred.append(poisson_model.predict(pd.DataFrame(data={'team': '{}'.format(j), 'opponent': '{}'.format(i),
                                         'home': 0}, index=[1]))[1])


buts_home_pred = pd.DataFrame(buts_home_pred, columns = ['Prédiction Buts Domicile'])
buts_away_pred = pd.DataFrame(buts_away_pred, columns = ['Prédiction Buts Extérieur'])
home_team = pd.DataFrame(home_team, columns=['HomeTeam'])
away_team = pd.DataFrame(away_team, columns=['AwayTeam'])
prob_domicile = pd.DataFrame(prob_domicile, columns=['Prédiction Domicile'])
prob_exterieur = pd.DataFrame(prob_exterieur, columns=['Prédiction Exterieur'])
prob_nuls = pd.DataFrame(prob_nuls, columns=['Prédiction Nuls'])
L1_1920_pred = pd.concat(
    [home_team, away_team, buts_home_pred, buts_away_pred, prob_domicile, prob_exterieur, prob_nuls], axis=1)
```

```python
L1_1920_final_pred = pd.merge(L1_1920_test, L1_1920_pred, on = ['HomeTeam', 'AwayTeam'])
L1_1920_final_pred
```

Nous convertissons les probabilités de victoire de chaque équipe en cotes.

```python
home_team = []
away_team = []
cotes_domicile = []
cotes_exterieur = []
cotes_nuls = []
for i, j in zip(L1_1920_test['HomeTeam'], L1_1920_test['AwayTeam']):
    match = simulate_match(poisson_model, '{}'.format(i),
                           '{}'.format(j), max_goals=10)
    home_team.append(i)
    away_team.append(j)
    cotes_domicile.append(1 / np.sum(np.tril(match, -1)))
    cotes_exterieur.append(1 / np.sum(np.triu(match, 1)))
    cotes_nuls.append(1 /np.sum(np.diag(match)))
    
home_team = pd.DataFrame(home_team, columns=['HomeTeam'])
away_team = pd.DataFrame(away_team, columns=['AwayTeam'])
cotes_domicile = pd.DataFrame(cotes_domicile, columns=['Prédiction Cote Domicile'])
cotes_nuls = pd.DataFrame(cotes_nuls, columns=['Prédiction Cote Nuls'])
cotes_exterieur = pd.DataFrame(cotes_exterieur, columns=['Prédiction Cote Exterieur'])
L1_1920_cotes_pred = pd.concat(
    [home_team, away_team, cotes_domicile,  cotes_nuls, cotes_exterieur], axis=1)
```

### Mise en place de la stratégie de paris

Nous récupérons les cotes du bookmaker Bet365 pour comparer les cotes.

```python
L1_1920_test_cotes = bookmaker[['HomeTeam', 'AwayTeam', 'B365H', 'B365D', 'B365A']][-10:]
L1_1920_test_cotes.reset_index(drop = True)
```

```python
L1_1920_final_cote = pd.merge(L1_1920_test_cotes, L1_1920_cotes_pred, on = ['HomeTeam', 'AwayTeam'])
L1_1920_final_cote = pd.concat([L1_1920_final_cote, prob_domicile, prob_exterieur, prob_nuls], axis = 1)
L1_1920_final_cote
```

Nous calculons notre espérance de gains (EV) en prenant en compte notre prédiction et la cote du bookmaker.

```python
L1_1920_final_cote['EV Domicile'] = 0 
L1_1920_final_cote['EV Extérieur'] = 0 
L1_1920_final_cote['EV Nuls'] = 0 

L1_1920_final_cote['EV Domicile'] = L1_1920_final_cote['B365H'] * (L1_1920_final_cote['Prédiction Domicile'] / 100)
L1_1920_final_cote['EV Extérieur'] = L1_1920_final_cote['B365A'] * (L1_1920_final_cote['Prédiction Exterieur'] / 100)
L1_1920_final_cote['EV Nuls'] = L1_1920_final_cote['B365D'] * (L1_1920_final_cote['Prédiction Nuls'] / 100)
```

Nous choissisons de parier sur un événement (Victoire, Nuls ou Défaite) que si notre EV est positive à priori.

```python
L1_1920_final_cote["Mise"] = 'Rien'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
    L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] < 1), 'Mise'] = 'Domicile'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (
    L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] < 1), 'Mise'] = 'Extérieur'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (
    L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] > 1), 'Mise'] = 'Nul'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
    L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] > 1), 'Mise'] = 'Domicile + Nul'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
    L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] < 1), 'Mise'] = 'Domicile + Extérieur'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (
    L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] > 1), 'Mise'] = 'Extérieur + Nul'
L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
    L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] > 1), 'Mise'] = 'Domicile + Extérieur + Nul'
```

### Conclusion du modèle de Poisson

Dans cette première partie, nous avons utilisé un modèle de Poisson pour prédire les matchs de Ligue 1. Nous sommes conscients que ce modèle présente de nombreuses limites et ne prend pas en compte tous les facteurs qui peuvent influer sur le résultat d'un match (par exemple : composition des équipes, forme / blessure des joueurs, dynamique de résultat des équipes, enchaînenement des matchs). Cependant, ce modèle prend en compte des caractéristiques essentielles dans la prédiction d'un match comme la force offensive de l'équipe, la qualité de l'opposition ainsi que l'avantage de jouer à domicile. Ainsi, ce modèle n'est bien entendu pas parfait et bien que nous concluons gagner de l'argent sur la prédiction de la 27 journée de championnat de Ligue 1, ce résultat est sans doute dû à de l'aléa statistiques plus qu'à une vraie performance prédictive de notre modèle. 
Dans la suite de ce projet, nous essayerons d'améliorer ce modèle en incluant des variables pertinentes au modèle. Nous essayerons également d'autres modèles que celui du modèle de Poisson pour prédire le score d'un match. Le premier modèle de Poisson, nous servira donc de benchmark pour juger de la performance de nos modèles "améliorés".

## Amélioration du modèle 

Les modèles présentés jusqu'à alors tentent de prédire le score d'un match de football. Une critique que nous pouvons faire à ce modèle concernent nos variables explicatives très restreintes. Nous proposons dans la suite de ce projet d'inclure les variables suivantes : 
  - **Indicateurs de qualité de l'équipe et des joueurs grâce aux notes de Fifa**. Les problèmes soulevés sont les suivants : 
      - *Inconnnue de l'adéquation des notes des joueurs au réél*. Ce point est bien entendu crucial dans notre analyse. Est-ce que les notes des joueurs Fifa reflètent-elles réellement le niveau des joueurs ? Nous ne pouvons pas apporter une réponse certaines à cette question tant les interprétations peuvent être sujets à de nombreuses controverses. Une chose est sûre, Fifa est le leader du marché des jeux vidéos sur la simulation de matchs de football. Les notes sont certainement attribuées selon un nombre de critères conséquent permettant de réfléter la force relative des joueurs. Dans notre situation, il ne nous est que peu important de savoir si l'attaquant du PSG Kylian Mbappé mérite vraiment sa note de 89/100. Ce qui est important est de savoir que l'attaquant du PSG est relativement meilleur que son comparse d'attaque Eric Choupo-Moting noté à 76/100. Il existe bien évidemment des situations où la qualité des joueurs est bien plus proche et où il est difficile par une note de dire lequel est le meilleur des deux. Nous sommes donc conscients des failles de l'approximation des joueurs par Fifa mais nous considérons que ces approximations sont acceptables et que Fifa reste une des meilleures sources possible dans l'approximation de la qualité des joueurs. 
      - *Distinction attaque/défense et intercheangeabilité des joueurs.* Fifa propose également une note agrégrée d'équipe (Général) ainsi qu'une moyenne par ligne de jeu (Défense / Milieu / Attaque). Nous récupérerons également ces données pour pallier ce problème. 
      - *Temps de jeu des joueurs et remplaçants*. Nous proposons dans un premier temps d'utiliser un onze "type" par équipe. L'idée à terme est d'ajuster notre modèle avant chaque rencontre dès que la composition des équipes pour chauqe journée est connue. Cela nécessaire un autre programme récupérant ces données. Cela est réalisable mais ne pourra as être mis en application étant donné qu'au moment où est réalisé ce projet les championnats de football sont suspendus en raison de l'épidémie de Coronavirus.
      

  - **Récupération des notes individuelles des joueurs attribuées par différentes sources de presse lors des matchs passés**. Ces données peuvent être intéressantes à récupérer si l'on utilise ces données comme un indicateur de la forme des joueurs sur une période donnée. En effet, ces données nous apporte une information sur un événement passé. En utilisant ces données de la manière suivante : moyenne des notes d'un joueur sur ces 5 derniers matchs, cela nous apporte une information sur la qualité de ses prestations sur les derniers matchs. Cet indicateur est différent des notes de Fifa car Fifa s'attache plus à la valeur intrinséque des joueurs alors que les notes des journaux sportifs donnent une note sur un moment précis. Un joueur ayant une très bonne note sur Fifa peut être dans une très mauvaise période et enchaîner les mauvais matchs tandis qu'un joueur moyen peut sur-performer et enchaîner les prestations abouties. Ainsi, la récupération des notes individuelles après chaque match peut apporter une information pertinente à l'analyse et complémentaire de celle déjà récupéré avec Fifa. Plusieurs questions apparraissent cependant quandt à la récupération de ces données de presse. En effet, les notes sont attribuées par un journaliste ou un panel de journalistes sportifs (ex : L'Equipe). Ces notes reflètent donc un avis subjectif sur la rencontre porté par un nombre restreint de personnes. De plus, ces personnes sont des journalistes et leur jugement ne peut donc pas être considéré comme un critère ultime de fiabilité. Il arrive régulièrement que des journalistes mettent une mauvaise note à un joueur alors même que l'entraîneur est satisfait du match de son joueur ou inversement. Les journalistes ne sont pas entraîneurs et ne regardent pas les actions de la même manière. Pour pallier à l'aspect subjectif des notes sur les journaux sportifs, il est possible de prendre la moyenne des notes de plusieurs journaux. La récupération des notes de plusieurs journaux est actuellement en cours d'études pour essayer de voir comment nous pouvons récupérer cet historique.
  
  
  - **Prise en compte d'évenements comme la modification du staff des équipes**. Cette information peut paraître pertinente à inclure bien que la littérature tend à montrer que changer d'entraîneur en cours de saisons n'impacte que de manière marginale les résultats de l'équipe. Ceci étant, même si cette information est marginale, elle peut tout de même améliorer le modèle et nous réflechirons à intégrer une telle variable. Une information qui présente un pouvoir explicatif et prédictif plus grand concerne la qualité des entraîneurs qui varie selon les équipes, au même titre que pour les joueurs. Il pourrait donc être intéressant de récupérer un indicateur de la qualité des entraîneurs. Les entraîneurs sont également notés après chaque match de championnat. Si nous parvenons à récupérer les informations sur les notes des joueurs, nous essayerons de récupérer également les notes attribuées aux entraîneurs. 
  
  
  - **Récupération de données météos et plus généralement des facteurs extérieurs entourant le match**. Les facteurs extérieurs entourant un match peuvent également impacter le cours d'un match. Le facteur extérieur le plus déterminant est sans doute constitué par les supporters et l'ambiance mise par ces derniers pour encourager leur équipe. Ce facteur extérieur est déjà pris en compte dans notre modèle grâce à la variable `home` qui permet de prendre cet avantage de jouer à domicile. Les autres facteurs extérieurs comme le jour, l'heure et la météo du match semblent des facteurs plus marginaux. Il n'y a en effet pas de raison évidente et intuitive pour lesquels une équipe soit plus efficace un samedi qu'un dimanche ou un jour ensolleilé plutôt qu'un jour de pluie. Ceci étant ces informations peuvent toujours apporté des informations à la marge et améliorer le modèle de quelques dizièmes de points de pourcentage. Nous garderons bien entendu toujours à l'esprit le risque d'overfitting présent avec un relativement petit échantillon.

```python
   def get_notes_lequipe_coach(url):
    
    home_team = []
    away_team = []
    score = []
    journee = []
    equipe_entraineur_domicile = []
    entraineur_domicile = []
    note_entraineur_domicile = []
    equipe_entraineur_exterieur = []
    entraineur_exterieur = []
    note_entraineur_exterieur = []
    arbitre = []
    note_arbitre = []


    for k in range(len(url)):
        try:
            get_page(url[k])

        except:
            pass

        try:
            journee.append(browser.find_element_by_class_name(
                "SelectNav__selectWrapper").text.rsplit(',', 2)[0].rsplit('\n', 2)[1])
            home_team.append(browser.find_element_by_class_name(
                "TeamScore").find_elements_by_tag_name('a')[0].text)
            away_team.append(browser.find_element_by_class_name(
                "TeamScore").find_elements_by_tag_name('a')[2].text)
            score.append(browser.find_element_by_class_name(
                "TeamScore").find_elements_by_tag_name('a')[1].text)
            entraineur_domicile.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__personName")[0].text)
            entraineur_exterieur.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__personName")[1].text)
            arbitre.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__personName")[2].text)
            equipe_entraineur_domicile.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__personInfo")[0].text)
            equipe_entraineur_exterieur.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__personInfo")[1].text)
            note_arbitre.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__ratingValue")[2].text)
            note_entraineur_domicile.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__ratingValue")[0].text)
            note_entraineur_exterieur.append(browser.find_elements_by_class_name(
                "RatingsArchivesPersonRating__ratingValue")[1].text)

        except:
            pass

    home_team = pd.DataFrame(home_team, columns=['HomeTeam'])
    away_team = pd.DataFrame(away_team, columns=['AwayTeam'])
    journee = pd.DataFrame(journee, columns=['Journée'])
    score = pd.DataFrame(score, columns=['Score'])
    entraineur_domicile = pd.DataFrame(
        entraineur_domicile, columns=['Entraineur Domicile'])
    equipe_entraineur_domicile = pd.DataFrame(
        equipe_entraineur_domicile, columns=['Equipe Coach Dom'])
    note_entraineur_domicile = pd.DataFrame(
        note_entraineur_domicile, columns=["Note Coach Dom"])
    entraineur_exterieur = pd.DataFrame(
        entraineur_exterieur, columns=['Entraineur Exterieur'])
    equipe_entraineur_exterieur = pd.DataFrame(
        equipe_entraineur_exterieur, columns=['Equipe Coach Ext'])
    note_entraineur_exterieur = pd.DataFrame(
        note_entraineur_exterieur, columns=["Note Coach Ext"])
    arbitre = pd.DataFrame(arbitre, columns=['Arbitre'])
    note_arbitre = pd.DataFrame(note_arbitre, columns=['Note Arbitre'])


    df_notes_lequipe_coach = pd.concat([journee, home_team, away_team, score, entraineur_domicile,
                                  equipe_entraineur_domicile, note_entraineur_domicile, entraineur_exterieur, 
                                  equipe_entraineur_exterieur, note_entraineur_exterieur, arbitre, note_arbitre], axis=1)
    df_notes_lequipe_coach.to_excel(
        "Données/Notes_Lequipe/df_notes_lequipe_coach.xls", index=False)
    
    return df_notes_lequipe_coach
```
