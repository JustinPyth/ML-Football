```python
def avantage_dom(df):

    df['Pts_HomeTeam'] = 0
    df['Pts_AwayTeam'] = 0

    df.loc[(df['HomeGoals'] > df['AwayGoals']), 'Pts_HomeTeam'] = 3
    df.loc[(df['HomeGoals'] > df['AwayGoals']), 'Pts_AwayTeam'] = 0
    df.loc[(df['HomeGoals'] == df['AwayGoals']), 'Pts_HomeTeam'] = 1
    df.loc[(df['HomeGoals'] == df['AwayGoals']), 'Pts_AwayTeam'] = 1
    df.loc[(df['HomeGoals'] < df['AwayGoals']), 'Pts_HomeTeam'] = 0
    df.loc[(df['HomeGoals'] < df['AwayGoals']), 'Pts_AwayTeam'] = 3

    domicile = df.groupby('HomeTeam')['Pts_HomeTeam'].mean()
    domicile = pd.DataFrame(domicile)
    domicile.reset_index(level=0, inplace=True)
    domicile = domicile.rename(columns={"HomeTeam":"Equipe", "Pts_HomeTeam":"Mean Pts Domicile"})

    exterieur = df.groupby('AwayTeam')['Pts_AwayTeam'].mean()
    exterieur = pd.DataFrame(exterieur)
    exterieur.reset_index(level=0, inplace=True)
    exterieur = exterieur.rename(columns={"AwayTeam":"Equipe", "Pts_AwayTeam":"Mean Pts Exterieur"})

    home_avantage = domicile.merge(exterieur, on ='Equipe')
    return home_avantage
```

```python
modele_buts_data = pd.concat([L1_1920_train[['HomeTeam','AwayTeam','HomeGoals']].assign(home=1).rename(
            columns={'HomeTeam':'team', 'AwayTeam':'opponent','HomeGoals':'goals'}),
           L1_1920_train[['AwayTeam','HomeTeam','AwayGoals']].assign(home=0).rename(
            columns={'AwayTeam':'team', 'HomeTeam':'opponent','AwayGoals':'goals'})])
```

```python
modele_buts_data.head()
```

```python
poisson_model = smf.glm(formula="goals ~ home + team + opponent", data=modele_buts_data, 
                        family=sm.families.Poisson()).fit()
print(poisson_model.summary())
```

L'intérprétation des coefficients d'un modèle de Poisson se fait de la même manière que pour un modèle de régression logistique, i.e on prend l'exponentielle du coefficient. Un coefficient positif ($e^{x}>1  \forall x > 0$) implique une probabilité de buts plus élevé tandis qu'un coefficient proche de 0 ($e^{0}=1$) indique un effet plus neutre. Prenons un exemple pour expliciter les coefficients. Prenons par exemple la dernière ligne de notre tableau de coefficients : 'home'. 'home' a un coefficient égal à 0.3477. Le signe positif du coefficient nous confirme l'intuition que l'on pouvait avoir que les équipes à domicile marquent en moyenne plus de buts que les équipes à l'extérieur. En particulier, c'est $e^{0.3477}=1.42$ fois plus probable que les équipes à domicile marquent plus de buts que les équipes à l'extérieur. 
D'autre part, nous voyons que le coefficient team associé au Paris SG est positif tandis que celui de Toulouse est négatif. Cela nous indique que le Paris SG (resp. Toulouse) marque en moyenne plus de buts (moins de buts) que la moyenne des équipes de Ligue 1. 
Enfin, les coefficients associés à opponent nous renseigne sur la force défensive des équipes. Nous voyons notamment que le coefficient associé au PSG (Toulouse) est négatif (positif) ce qui signifie que le PSG encaisse en moyenne relativement moins de buts (plus de buts) que les autres équipes du championnat. 

### Prédiction : Exemple sur un match

```python
def simulate_match(modele_foot, EquipeDom, EquipeExt, max_goals=10):
    
    home_goals_avg = modele_foot.predict(pd.DataFrame(data={'team': EquipeDom,
                                                           'opponent': EquipeExt, 'home': 1},
                                                     index=[1])).values[0]
    away_goals_avg = modele_foot.predict(pd.DataFrame(data={'team': EquipeExt,
                                                           'opponent': EquipeDom, 'home': 0},
                                                     index=[1])).values[0]
    team_pred = [[poisson.pmf(i, team_avg) for i in range(
        0, max_goals+1)] for team_avg in [home_goals_avg, away_goals_avg]]
    
    return(home_goals_avg, away_goals_avg, np.outer(np.array(team_pred[0]), np.array(team_pred[1])))
```

```python
def betting(site):
    df_bookie = bookmaker[['HomeTeam', 'AwayTeam', '{}H'.format(
        site), '{}D'.format(site), '{}A'.format(site)]][-19:]
    df_bookie = df_bookie.reset_index(drop=True)

    L1_1920_final_cote = pd.merge(df_bookie, L1_1920_cotes_pred, on=[
                                  'HomeTeam', 'AwayTeam'])
    L1_1920_final_cote = pd.concat(
        [L1_1920_final_cote, prob_domicile, prob_exterieur, prob_nuls], axis=1)

    L1_1920_final_cote['EV Domicile'] = 0
    L1_1920_final_cote['EV Extérieur'] = 0
    L1_1920_final_cote['EV Nuls'] = 0

    L1_1920_final_cote['EV Domicile'] = L1_1920_final_cote['{}H'.format(
        site)] * (L1_1920_final_cote['Prédiction Domicile'] / 100)
    L1_1920_final_cote['EV Extérieur'] = L1_1920_final_cote['{}A'.format(
        site)] * (L1_1920_final_cote['Prédiction Exterieur'] / 100)
    L1_1920_final_cote['EV Nuls'] = L1_1920_final_cote['{}D'.format(
        site)] * (L1_1920_final_cote['Prédiction Nuls'] / 100)

    L1_1920_final_cote["Mise"] = 'Rien'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
        L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] < 1), 'Mise'] = 'Domicile'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (
        L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] < 1), 'Mise'] = 'Extérieur'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (
        L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] > 1), 'Mise'] = 'Nul'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (L1_1920_final_cote['EV Extérieur'] > 1) & (
        L1_1920_final_cote['EV Nuls'] > 1) & (L1_1920_final_cote['EV Extérieur'] > L1_1920_final_cote['EV Nuls']), 'Mise'] = 'Extérieur'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] < 1) & (L1_1920_final_cote['EV Extérieur'] > 1) & (
        L1_1920_final_cote['EV Nuls'] > 1) & (L1_1920_final_cote['EV Extérieur'] < L1_1920_final_cote['EV Nuls']), 'Mise'] = 'Nul'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
        L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] > 1) & (L1_1920_final_cote['EV Domicile'] > L1_1920_final_cote['EV Nuls']), 'Mise'] = 'Domicile'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
        L1_1920_final_cote['EV Extérieur'] < 1) & (L1_1920_final_cote['EV Nuls'] > 1) & (L1_1920_final_cote['EV Domicile'] < L1_1920_final_cote['EV Nuls']), 'Mise'] = 'Nul'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
        L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] < 1) & (L1_1920_final_cote['EV Domicile'] < L1_1920_final_cote['EV Extérieur']), 'Mise'] = 'Extérieur'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (
        L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] < 1) & (L1_1920_final_cote['EV Domicile'] > L1_1920_final_cote['EV Extérieur']), 'Mise'] = 'Domicile'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] > 1) & ((L1_1920_final_cote['EV Domicile'] > L1_1920_final_cote['EV Extérieur']) & (L1_1920_final_cote['EV Domicile'] > L1_1920_final_cote['EV Nuls'])), 'Mise'] = 'Domicile'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] > 1) & ((L1_1920_final_cote['EV Extérieur'] > L1_1920_final_cote['EV Domicile']) & (L1_1920_final_cote['EV Extérieur'] > L1_1920_final_cote['EV Nuls'])), 'Mise'] = 'Extérieur'
    L1_1920_final_cote.loc[(L1_1920_final_cote['EV Domicile'] > 1) & (L1_1920_final_cote['EV Extérieur'] > 1) & (L1_1920_final_cote['EV Nuls'] > 1) & ((L1_1920_final_cote['EV Nuls'] > L1_1920_final_cote['EV Domicile']) & (L1_1920_final_cote['EV Nuls'] > L1_1920_final_cote['EV Extérieur'])), 'Mise'] = 'Nul'

    L1_1920_test['Résultat'] = 'Res'
    L1_1920_test.loc[(L1_1920_test['HomeGoals'] >
                      L1_1920_test['AwayGoals']), 'Résultat'] = 'Domicile'
    L1_1920_test.loc[(L1_1920_test['HomeGoals'] <
                      L1_1920_test['AwayGoals']), 'Résultat'] = 'Extérieur'
    L1_1920_test.loc[(L1_1920_test['HomeGoals'] ==
                      L1_1920_test['AwayGoals']), 'Résultat'] = 'Nul'

    L1_1920_final_cote['Résultat'] = L1_1920_test['Résultat'].values

    L1_1920_final_cote['Score Prédiction']=''
    L1_1920_final_cote.loc[(L1_1920_final_cote['Mise'] == L1_1920_final_cote['Résultat']) , 'Score Prédiction'] = 'Bonne'
    L1_1920_final_cote.loc[(L1_1920_final_cote['Mise'] != L1_1920_final_cote['Résultat']) , 'Score Prédiction'] = 'Mauvaise'
    L1_1920_final_cote.loc[(L1_1920_final_cote['Mise'] == 'Rien') , 'Score Prédiction'] = 'Rien'

    L1_1920_final_cote['Gain ou Perte'] = 0
    L1_1920_final_cote.loc[(L1_1920_final_cote['Score Prédiction']
                        == 'Rien'), 'Gain ou Perte'] = 0
    L1_1920_final_cote.loc[(L1_1920_final_cote['Score Prédiction']
                        == 'Mauvaise'), 'Gain ou Perte'] = -10
    L1_1920_final_cote.loc[(L1_1920_final_cote['Score Prédiction'] == 'Bonne') &
                       (L1_1920_final_cote['Mise'] == 'Domicile'), 'Gain ou Perte'] = 10*L1_1920_final_cote['{}H'.format(site)] - 10
    L1_1920_final_cote.loc[(L1_1920_final_cote['Score Prédiction'] == 'Bonne') &
                       (L1_1920_final_cote['Mise'] == 'Extérieur'), 'Gain ou Perte'] = 10*L1_1920_final_cote['{}A'.format(site)] - 10
    L1_1920_final_cote.loc[(L1_1920_final_cote['Score Prédiction'] == 'Bonne') &
                       (L1_1920_final_cote['Mise'] == 'Nul'), 'Gain ou Perte'] = 10*L1_1920_final_cote['{}D'.format(site)] - 10
    
    print('En suivant les prédictions de notre modèle et en pariant 10€ sur le site {} sur chaque match nous aurions fait {}€ de gains sur la journée de championnat.'.format(
    site, round(L1_1920_final_cote['Gain ou Perte'].sum()), 2))
    
    return L1_1920_final_cote
```

```python
FP = cm.sum(axis=0) - np.diag(cm) 
FN = cm.sum(axis=1) - np.diag(cm)
TP = np.diag(cm)
TN = cm.sum() - (FP + FN + TP)
```

```python
# Sensitivity, hit rate, recall, or true positive rate
TPR = TP/(TP+FN)
# Specificity or true negative rate
TNR = TN/(TN+FP) 
# Precision or positive predictive value
PPV = TP/(TP+FP)
# Negative predictive value
NPV = TN/(TN+FN)
# Fall out or false positive rate
FPR = FP/(FP+TN)
# False negative rate
FNR = FN/(TP+FN)
# False discovery rate
FDR = FP/(TP+FP)

# Overall accuracy
ACC = (TP+TN)/(TP+FP+FN+TN)
```
